```python
def get_audio_length(file_path):
  audio = MP3(file_path)
  return audio.info.length
```

```python
err = []
err_messages = []
```

```python
# genres = ['Gospel', 'Highlife', 'HipHop Asakaa', 'Love Poem', 'Rap', 'Soul', 'Traditional Music']
genre = 'Highlife'
```

```python
norm_audios_feat = []
count = 1
#metadata = pd.read_csv("../CSV_Files/AudioSegments.csv",encoding='latin1')
#metadata = pd.read_csv("AudioSegments.csv",encoding='latin1') # Audio Segments_Highlife_Traditional song.csv
# file_path = 'gdrive/MyDrive/Multimodal Ghana Music Videos/Dataset/Audio Segments_Highlife_Traditional.csv'
csv_path = f'gdrive/MyDrive/Multimodal Ghana Music Videos/Dataset/{genre}/{genre}.csv'
base_path = 'gdrive/MyDrive/Multimodal Ghana Music Videos/'
metadata = pd.read_csv(csv_path, encoding='latin1')
for index_num,row in tqdm(metadata.iterrows()):
    try:
      file = metadata.loc[index_num, "Filename"].strip()
      link = metadata.loc[index_num, "URL"]
      title = metadata.loc[index_num, "Title of Song"]
      lang = metadata.loc[index_num, "Language"]
      sent = metadata.loc[index_num, "Sentiment"]
      text = metadata.loc[index_num, "Lyrics"]
      gen = metadata.loc[index_num, "Genre"]
      meaning = metadata.loc[index_num, "Meaning of Song"]
      #y , sr = librosa.load(file)
      file_path = f'{base_path}{genre}/{title}/{file}'
      # print('file_path', file_path, '\n')
      y, sr = librosa.load(file_path, res_type='kaiser_fast', duration=get_audio_length(file_path), offset=0.5, sr=44100)
      #feat_vect_mfcc = [np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)]
      mfccs_features = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
      #feat_vect_mfcc = [np.mean(librosa.feature.mfcc(y=y, sr=sr))]
      feat_vect_iii = [np.mean(librosa.onset.onset_strength(y=y, sr=sr))]
      feat_vect_iiii = [np.mean(librosa.feature.chroma_stft(y=y, sr=sr))]
      feat_vect_iiiii = [np.mean(librosa.effects.hpss(y))]
      feat_vect_ii = [np.mean(librosa.feature.zero_crossing_rate(y))]
      mfccs_scaled_features = [np.mean(mfccs_features.T,axis=0)]
      feat_vect_first = [count]
      feat_vect_second = [file]
      one = [link]
      feat_vect_third = [title]
      two = [lang]
      feat_vect_last = [sent]
      three = [text]
      four = [gen]
      five = [meaning]
      #feat_vect_centroid = [np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))]
      feature_vector = feat_vect_first + feat_vect_second + feat_vect_third + one + two + three + four + five + feat_vect_last + feat_vect_iii + feat_vect_iiii + feat_vect_iiiii + feat_vect_ii + mfccs_scaled_features
      #feature_vector = feat_vect_first + feat_vect_second + feat_vect_third + mfccs_scaled_features + feat_vect_last
      norm_audios_feat.append(feature_vector)
      count = count + 1
    except Exception as error_message:
        err.append(str(index_num))
        err_messages.append((index_num, str(error_message)))
    # if index_num == 10:
    #   break
```

```python
save_path = f'{base_path}Dataset/{genre}'
norm_output = f'{save_path}/{genre}_Acoustic_Features_00.csv'

# norm_output = 'Acoustic_Features_00.csv'
# norm_path
header =[
 'SN',
 'Filename',
 'Title of Song',
 'Url',
 'Language',
 'Lyrics',
 'Genre',
 'Meaning of Song',
 'Sentiment',
 'Onset_Strength',
 'Chroma_Shift',
 'HPSS',
 'Zero_Crossing_Rate',
 'MFCC'
]


with open(norm_output,'+w',encoding='utf-8') as f:
 csv_writer = csv.writer(f, delimiter = ',',quoting=csv.QUOTE_MINIMAL)
 csv_writer.writerow(header)
 csv_writer.writerows(norm_audios_feat)
```

```python
# csv_path = 'gdrive/MyDrive/Multimodal Ghana Music Videos/Dataset/Audio Segments_Highlife_Traditional song.csv'
# base_path = 'gdrive/MyDrive/Multimodal Ghana Music Videos/'
# save_path = f'{base_path}Dataset/{genre}'
# acoustic_path = f'{save_path}/{genre}_Acoustic_Features_00.csv'
final_acoustic_path = f'{save_path}/{genre}_Final_Acoustic_Features.csv'

with open(norm_output) as input, open(final_acoustic_path, 'w') as output:
    non_blank = (line for line in input if line.strip())
    output.writelines(non_blank)


with open(f'{save_path}/{genre}_logs', 'w') as error_file:
    error_str = ', '.join(err)
    print(error_str)
    error_file.write(f'{error_str}, \n')
    for err in err_messages:
        error_file.write(f"Index: {str(err[0])}, Error: {err[1]} \n")
```
