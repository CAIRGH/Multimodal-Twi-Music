```python
dataset_path = 'gdrive/MyDrive/Multimodal Ghana Music Videos/Dataset/'
# !ls 'gdrive/MyDrive/Multimodal Ghana Music Videos/Dataset/'
```

```python
# Load the CSV file into a pandas DataFrame
csv_file = dataset_path + "segments_length.csv"  # Replace with your CSV file path
data = pd.read_csv(csv_file)

# Ensure the required columns exist
# Plot the bar chart
genre = data['Genre']
segments = data['Segments']
plt.figure(figsize=(12, 10))  # Set the figure size
plt.bar(genre, segments, color='skyblue')

# Customize the chart
plt.title("Distribution of Music Genre", fontsize=16)
plt.xlabel("Number of Video Segments", fontsize=12)
plt.ylabel("Music Genre", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: Add a grid

output_file = dataset_path + "graphs/genre_distribution.pdf"  # Change file name or format as needed
plt.savefig(output_file)  # High-quality save


# Show the plot
# plt.tight_layout()
plt.show()
```

```python
audio_segments = dataset_path + 'Audio Segments.csv' # "./data/Audio Segments.csv"
audio_data = pd.read_csv(audio_segments)
audio_data = audio_data.drop(columns=['Unnamed: 10',
       'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14',
       'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
       'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22',
       'Unnamed: 23', 'Unnamed: 24'])
# audio_data = pd.DataFrame(audio_data)
print('columns', audio_data.columns)
```

```python
aggregated_language = audio_data.groupby('Language').count().reset_index()[['Language', 'SN']]
print(aggregated_language)
```

```python
# Plot the bar chart
language = aggregated_language['Language']
segments_lang = aggregated_language['SN']
plt.figure(figsize=(12, 10))  # Set the figure size
plt.bar(language, segments_lang, color='skyblue')

# Customize the chart
plt.title("Distribution of Languages in Music Segments", fontsize=16)
plt.xlabel("Number of Video Segments", fontsize=12)
plt.ylabel("Language", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: Add a grid

output_file = dataset_path + "graphs/language_distribution.pdf"  # Change file name or format as needed
plt.savefig(output_file)  # High-quality save


# Show the plot
# plt.tight_layout()
plt.show()
```

```python
colors = ['']

# Create pie chart
plt.pie(segments_lang, labels=language,   autopct=lambda p: f'{int(p * sum(aggregated_language["SN"]) / 100)}',\
        startangle=90, colors=plt.cm.Set2.colors)


# Add title
plt.title('Pie Chart of Distribution of Languages in Music Segments')

output_file = dataset_path + "graphs/language_distribution_pie.pdf"  # Change file name or format as needed
plt.savefig(output_file)  # High-quality save

# Display the chart
plt.show()
```

```python
aggregated_sentiment = audio_data.groupby('Sentiment').count().reset_index()[['Sentiment', 'SN']]
print(aggregated_sentiment)
```

```python
# Plot the bar chart
sentiment = aggregated_sentiment['Sentiment']
segments_sent = aggregated_sentiment['SN']
plt.figure(figsize=(12, 10))  # Set the figure size
plt.bar(sentiment, segments_sent, color='skyblue')

# Customize the chart
plt.title("Distribution of Sentiment in Music Segments", fontsize=16)
plt.xlabel("Number of Video Segments", fontsize=12)
plt.ylabel("Sentiment", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: Add a grid

output_file = dataset_path + "graphs/sentiment_distribution.pdf"  # Change file name or format as needed
plt.savefig(output_file)  # High-quality save


# Show the plot
# plt.tight_layout()
plt.show()
```

```python
# Create pie chart
plt.pie(segments_sent, labels=sentiment, autopct=lambda p: f'{int(p * sum(aggregated_sentiment["SN"]) / 100)}',\
        startangle=90, colors=plt.cm.Set2.colors)

# Add title
plt.title('Pie Chart of Distribution of Languages in Music Segments')

output_file = dataset_path + "graphs/sentiment_distribution_pie.pdf"  # Change file name or format as needed
plt.savefig(output_file)  # High-quality save

# Display the chart
plt.show()
```
