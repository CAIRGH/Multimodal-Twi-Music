```python
def split_audio(input_video_path, output_folder=None):
    # print('input_video_path', f' {input_video_path} \n')

    if output_folder is None:
        output_folder = os.path.dirname(input_video_path)

    os.makedirs(output_folder, exist_ok=True)

    name_without_extension = os.path.splitext(os.path.basename(input_video_path.split('/')[-1]))[0]
    output_audio_path = os.path.join(output_folder, f"{name_without_extension}.mp3")
    clip = mp.VideoFileClip(input_video_path)
    clip.audio.write_audiofile(output_audio_path)
```
