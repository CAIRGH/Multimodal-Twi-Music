```python
# Function to split a single video into chunks
def split_video_ffmpeg(input_video_path, chunk_duration=15, output_folder=None):
    # Get the video durationinput_video_path
    video = VideoFileClip(input_video_path)
    video_duration = video.duration

    # Use a default output folder if none is provided
    if output_folder is None:
        output_folder = f"{os.path.splitext(os.path.basename(input_video_path))[0]}"

    # Create the output folder
    os.makedirs(output_folder, exist_ok=True)

    # Split the video
    start_time = 0
    chunk_number = 1

    while start_time < video_duration:
        end_time = min(start_time + chunk_duration, video_duration)
        name_without_extension = output_folder.split('/')[-1]#os.path.splitext(os.path.basename(output_folder))[0]
        print('name_without_extension', f' \n {name_without_extension} \n')
        output_video_path = os.path.join(output_folder, f"Segment_{chunk_number}_{name_without_extension}.mp4")

        # Extract the chunk
        ffmpeg_extract_subclip(input_video_path, start_time, end_time, targetname=output_video_path)

        start_time = end_time
        chunk_number += 1

    print(f"Video '{input_video_path}' has been split into {chunk_number - 1} chunks in folder '{output_folder}'.")

# Process all .mp4 files in a folder
def split_all_videos_in_folder(folder_path, chunk_duration=15, output_folder=None):
    # if not os.path.isdir(folder_path):
    #     raise NotADirectoryError(f"The provided path '{folder_path}' is not a directory.")

    for file_name in os.listdir(folder_path):
        if file_name.endswith(".mp4"):
            input_video_path = os.path.join(folder_path, file_name)
            print(f"Processing: {input_video_path}")
            # Create a subfolder for each video if output_folder is None
            video_output_folder = output_folder or os.path.join(folder_path, f"{os.path.splitext(file_name)[0]}")
            split_video_ffmpeg(input_video_path, chunk_duration, video_output_folder)
```
